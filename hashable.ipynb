```python
# 이렇게 같은 내용을 가지는 object들은 id들이 전부 같다.

id([1, 2, 3]) == id([1, 2, 3])
```

```python
# list는 hashable 하지 않다.

hash([1, 2, 3])
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

id(a) == id(b)
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
print(type(a))
# 다음의 답은 뭘까? 
a == b

# False이다.
# 왜 False일까?
# 왜냐하면 a가 참조하는 [1, 2, 3]와 b 가 참조하는 [1, 2, 3]이 다른 인스턴스이기 때문이다.

```

```python
a = (1, 2, 3)
b = (1, 2, 3)


a == b
# 이거의 답은 무엇일까?

# True

# 튜플은 hashable한 클래스다.
# 튜플 클래스의 인스턴스를 생성할 때 넘겨지는 파라미터가 같으면 항상 완전히 같은 인스턴스가 생성된다. (id)

# immutable 하면 hashable하다. 역은 성립하지 않는다.
# tuple, string, integer, booleans 클래스의 인스턴스는 immutable 하고 hashable 하다.
```
